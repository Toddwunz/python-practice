```python
def maskify(cc):
    lst = []
    for i in cc[0:-4]:
        lst.append('#')
    lst.append(cc[-4:])
    return ''.join(lst)
```

```python
MORSE_CODE = {'....':'H', '.':'E','-.--':'Y'}
```

```python
def Morsecode(char):
    return MORSE_CODE[char]
```

```python
words = re.split('  +',line)
for item in words:
    chars = re.split(' ',item)
    r = map(Morsecode,chars)
    print(''.join(list(r)))
```
